```python
Number_struck = df['Wildlife: Number struck'].value_counts()
Number_struck
```

```python
plt.figure(figsize=(5, 4))
sns.barplot(x=Number_struck.index, y=Number_struck,color="green")
plt.title('Wildlife: Number struck')
plt.xlabel('Interval')
plt.ylabel('Number struck')
plt.show()
```

```python
yearly_strikes = df['Year'].value_counts().sort_index(ascending=True)
yearly_strikes
```

```python
# Plotting a line chart
plt.figure(figsize=(8, 6))  # Adjust figure size if needed
plt.plot(yearly_strikes.index,yearly_strikes, marker='o', linestyle='-', color='b', label='Sales')  # Line plot
plt.title('Strikes Over Years')  # Adding title to the plot
plt.xlabel('Year')  # Adding label to the x-axis
plt.ylabel('Strike count')  # Adding label to the y-axis
plt.grid(True)  # Adding grid lines for better readability
plt.legend()  # Showing legend
plt.show()
```

```python
strikes_by_altitude = df['Altitude bin'].value_counts()

plt.figure(figsize=(3, 2))
sns.barplot(x=strikes_by_altitude.values, y=strikes_by_altitude.index,color="skyblue")
plt.title('Bird Strikes by Altitude')
plt.xlabel('Number of Strikes')
plt.ylabel('Altitude Bin')
plt.show()
```

```python
Airport_name = df["Airport: Name"].value_counts().head(8)
Airport_name
```

```python
#Yearly Cost Incurred due to Bird Strikes
yearly_cost=df.groupby(df["Year"])["Cost: Total $"].sum()
yearly_cost
```

```python
# Plotting a vertical bar graph
plt.figure(figsize=(6, 2))  
plt.bar(yearly_cost.index, yearly_cost, color='violet')  
plt.title('Yearly Cost Analysis') 
plt.xlabel('Year')  
plt.ylabel('Cost') 

plt.show()
```

```python
#Were Pilots Informed? & Prior Warning and Effect of Strike Relation
warning_effect = df.groupby(df["Pilot warned of birds or wildlife?"])["Effect: Indicated Damage"].count()
warning_effect
```

```python
# Plotting a vertical bar graph
plt.figure(figsize=(6, 2))  
plt.bar(warning_effect.index, warning_effect, color='violet')  
plt.title('Prior Warning and Effect of Strike Relation') 
plt.xlabel('Year')  
plt.ylabel('Cost') 

plt.show()
```
